### 프로젝트 목표 및 내용

저는 금융관리 중 주식에 대해서 관심이 풍부합니다. 매번 주식은 시세가 한 곳으로 치우쳐 있지 않고 어떨때는 올라갔다가 또 어떨 때는 내려가는 경우를 많이 합니다. 주식에 대한 정보를 풍부하게 알아야 만 주식으로 돈을 벌 수 있거나 돈을 말아먹는 경우로 결정한다고 합니다. 하지만, 이 과정에서 주식이라는 데이터가 과연 어떠한 것을 활용하여서 주식데이터를 저희에게 제공하는지도 궁금합니다. 그래서 저희는 주식데이터를 이용하여서 정보를 입력하고 앞으로 주식이 떨어질지 올라갈지 파악하는 것을 목표로 합니다.<br>
이 프로젝트는 주식을 해왔던 기록들을 혼자가 아닌 개발자가 이용하는 시스템을 이용하여서 모으려고 합니다. 이 데이터들을 이용하여 주식데이터 기록들을 한눈에 알아볼 수 있도록 그래프, 표와 같은 자료를 이용하여서 나타내도록 할 것입니다. 이를 통해 주식 종목을 입력하면 정보와 시세들을 자동으로 할 수 있도록 하는 프로젝트입니다. 

###  주제 선정 및 이유 또는 프로젝트의 필요성

저는 주식에 대한 시세와 현황을 보는 것이 관심 있습니다. 주식은 지금까지 매수와 매도를 한 데이터로 자신들이 정보들을 모아서 지금 주식현황이 어떠한지 결정하는 것입니다. 그래서 저희는 이러한 주식종목들을 데이터로 많은 사람들에게 주식항목을 검색하면 자동적으로 코스피, 시종 단가와 같은 세세한 항목들을 알려줄려고 하기 위해 이러한 주제를 선정하게 되었습니다. 주식이라는 것은 우리가 여태까지 사거나 팔았던 기록들 중 한 부분입니다. 그래서 신문, 뉴스, 인터넷 증권 홈페이지와 같은 매체와 같이 저는 지금 현 시점에서 주식이 어떻게 돌아가고 있는가에 대한 데이터를 파악을 합니다 . 또한 주식이라는 것은 유동적인 항목들 중 하나이기 때문에 주식에 대한 순위정보이나 흑자, 적자 정보들이 있으면 사람들이 물 흐르듯이 매수를 하든가 매도를 하게 됩니다.<br>
이와 같은 주식 프로젝트를 통해 사람들이 어떤 시기에 주식에 대한 정보를 정확하게 분석을 하는 것, 분석을 통해 자신이 무엇을 해야 하는지에 계획하는 것, 그리고 주식을 매수를 해야 하는 것과 주식을 매도하는 것을 통해 주식을 해왔던 것으로 사람들에게 주식 정보를 줄 수 있도록 할 것입니다.

### 데이터 획득

일단 저는 주식 데이터들을 수집하기 위해서는 Open API를 이용하여 획득할 생각입니다. 일단 Open API라는 것은 API서비스 중에서 입력코드를 만드는 것 없이 마음대로 서비스를 등록할 수 있는 것을 말합니다. Open API를 이용하면은 실시간으로 주식의 현재 상황을 그래프와 차트 표 등으로 나타낼 수 있도록 하고 저희가 코드에 명령을 입력하면 저절로 이에 관련된 자료가 자동적으로 보여줍니다. 예를 들자면 주식 종목 중 실시간 시세를 알아보겠다고 명령하면 Open API가 이를 자동적으로 수행합니다. 이로써 제가 원하는 항목과 그 데이터가 나오면서 앞으로 어떠한 전망이 보일지 파악할 수 있기 때문입니다.<br>
Open API를 이용한 결과 주식에 대한 항목들이 나오게 되면서 항목을 선택하면 선택한 항목에 대한 정보가 다 나오도록 되어 있는 시스템으로 작동될 것이라고 예상합니다. 그 주식 홈페이지들 중 키움증권이라는 사이트가 있습니다. <br>
그리고 OPEN API는 개발환경이더라도 파이썬 언어를 직접적으로 접근하기 힘듭니다. 왜냐하면 API 오픈 소스가 C#이기 때문에 직접적으로는 C# 언어만 가능합니다. 하지만, 직접적이지는 않더라도 명령연결통로를 수행하면 파이썬으로 수행할 수 있다고 하는데 그건 바로 KOA STUDIO에 나온 관련함수 데이터를 기반으로 접근할 생각입니다. 여기에 관련 함수와 관련된 목록들이 자세하게 나타나져 있습니다. 이 명령어를 이용해서 데이터들을 활용할 생각입니다.<br>
이 사이트는 매번 실시간으로 주식, 채권, LTF와 같은 금융에 관려된 데이터를 실시간으로 어떤 현황인지 알 수 있는 홈페이지입니다. 저는 이 키움증권 Open API를 이용하여 지금까지 실시간 주식 현황에 대한 것을 추출할 것입니다. 실시간 주식 시세를 파악하면은 주식의 데이터들을 확실히 알아낼 것 입니다. 물론 네이버 증권사 홈페이지들도 있지만, 네이버 증권사 홈페이지들은 실시간으로 주식을 알 수 있는 홈페이지가 아니기 때문에 데이터 획득으로 이용할 때에는 주식 실시간 홈페이지 중 키움증권 OpenAPI를 이용하는 것이 더 낫다고 생각하여서 주식 데이터 획득을 할 생각입니다. 그리고 데이터 획득하기 전에 주식 모의투자를 신청해야 한다고 들었습니다. 일단 주식이라는 것이 한번 잘못 매수/매도하면은 금융적인 타격이 크기 때문에 키움증권 홈페이지에서 실시간 투자가 아닌 주식 모의 투자를 먼저 해볼 생각입니다. 키움증권에 주식 모의 투자를 신청하면은 자동적으로 Open API 접속 서버로 들어가서 제가 원하는 데이터 항목에 검색하는 활동으로 데이터를 회득할 생각입니다.

또한 주식들 중 주식 종목에 대해 알고 싶은 내용과 주식 종목을 선택하면 이러한 것을 데이터로 일봉, 주봉, 년봉 데이터별로 조회 할 수 있다는 것입니다.<br>
이와 같은 정보를 보여주면서 어떠한 데이터를 갖고 있고 그 데이터로 무슨 분야를 만들고 있는지 그 주식 종목에 대한 데이터를 세세하게 알아볼 수 있도록 할 것입니다. 그러면 자기가 원하는 항목을 누르면 제가 원하는 데이터(정보)들을 확인하는 것입니다. 이를 통해 제가 원하는 Open API를 구현해낼 것입니다.이를 통해  언제 어디에서나 우리가 데이터를 관찰함에 따라서 그 주식에 대한 흑자와 적자를 구문지을 것입니다. 그 정보를 이용하여서 정보를 실시간으로 어떻게 돌아가고 있는지 데이터를 저장할 수 있고, 저희들이 그 데이터들을 파악할 수 있게 해주는 것입니다.<br>

지금 이것은 일봉데이터추출 그리고 저장할 수 있는 기능입니다. 이것도 마찬가지로 class 기능을 이용합니다.

```python
TR_REQ_TIME_INTERVAL = 0.2

class Kiwoom(QAxWidget):
    def __init__(self):
        super().__init__()
        self._create_kiwoom_instance()
        self._set_signal_slots()
        
    # KIWOOM OPEN API 실행
    def _create_kiwoom_instance(self):
        self.setControl("KHOPENAPI.KHOpenAPICtrl.1")
        
    # tr목록을 세팅
    def _set_signal_slots(self):
        self.OnEventConnect.connect(self._event_connect)
        self.OnReceiveTrData.connect(self._receive_tr_data)
    
    # 로그인이 되었는지 수행
    def comm_connect(self):
        self.dynamicCall("CommConnect()")
        self.login_event_loop = QEventLoop()
        self.login_event_loop.exec_()

    def _event_connect(self, err_code):
        if err_code == 0:
            print("connected")
        else:
            print("unconnected")

        self.login_event_loop.exit()
    # 종목코드 추출
    def get_code_list_by_market(self, market):
        code_list = self.dynamicCall("GetCodeListByMarket(QString)", market)
        code_list = code_list.split(';')
        return code_list[:-1]
    
    # 종목코드에 있는 주식회사를 한글로 입력
    def get_master_code_name(self, code):
        code_name = self.dynamicCall("GetMasterCodeName(QString)", code)
        return code_name

    def set_input_value(self, id, value):
        self.dynamicCall("SetInputValue(QString, QString)", id, value)
        
    # 데이터 추출
    def comm_rq_data(self, rqname, trcode, next, screen_no):
        self.dynamicCall("CommRqData(QString, QString, int, QString)", rqname, trcode, next, screen_no)
        self.tr_event_loop = QEventLoop()
        self.tr_event_loop.exec_()

    def _get_comm_data(self, trcode, record_name, index, item_name):
        ret = self.dynamicCall("GetCommData(QString, QString, int, QString)", trcode,
                               record_name, index, item_name)
        return ret.strip()

    def _get_repeat_cnt(self, trcode, rqname):
        ret = self.dynamicCall("GetRepeatCnt(QString, QString)", trcode, rqname)
        return ret

    def _receive_tr_data(self, screen_no, rqname, trcode, record_name, next, unused1, unused2, unused3, unused4):
        if next == '2':
            self.remained_data = True
        else:
            self.remained_data = False

        if rqname == "opt10081_req":
            self._opt10081(rqname, trcode)

        try:
            self.tr_event_loop.exit()
        except AttributeError:
            pass
        
    # 일봉데이터들을 추출하여서 데이터 분석 및 저장
    def _opt10081(self, rqname, trcode):
        data_cnt = self._get_repeat_cnt(trcode, rqname)

        for i in range(data_cnt):
            TR_code = self._get_comm_data(trcode, rqname, i,"종목코드")
            volume = self._get_comm_data(trcode, rqname, i, "거래량")
            open = self._get_comm_data(trcode, rqname, i,"시가")
            high = self._get_comm_data(trcode, rqname, i, "고가")
            low = self._get_comm_data(trcode, rqname, i, "저가")
            close = self._get_comm_data(trcode, rqname, i, "현재가")
            print(TR_code, volume + "원" + open + "원" + high + "원" + low + "원" + close + "원")

            self.ohlcv['TR_code'].append(TR_code)
            self.ohlcv['volume'].append(int(volume))
            self.ohlcv['open'].append(int(open))
            self.ohlcv['high'].append(int(high))
            self.ohlcv['low'].append(int(low))
            self.ohlcv['close'].append(int(close))
            

if __name__ == "__main__":
    app = QApplication(sys.argv)
    kiwoom = Kiwoom()
    kiwoom.comm_connect()
    kiwoom.ohlcv = {'TR_code': [], 'volume': [], 'open': [], 'high': [], 'low': [], 'close': []}

    # opt10081 TR 요청
    kiwoom.set_input_value("종목코드", "039490")
    kiwoom.set_input_value("기준일자", "20170224")
    kiwoom.set_input_value("수정주가구분", 1)
    kiwoom.comm_rq_data("opt10081_req", "opt10081", 0, "0101")

    while kiwoom.remained_data == True:
        time.sleep(TR_REQ_TIME_INTERVAL)
        kiwoom.set_input_value("종목코드", "039490")
        kiwoom.set_input_value("기준일자", "20170224")
        kiwoom.set_input_value("수정주가구분", 1)
        kiwoom.comm_rq_data("opt10081_req", "opt10081", 2, "0101")
        
    # pandas를 이용하여서 DB Database SQLITE에다가 저장
    df = pd.DataFrame(kiwoom.ohlcv, columns=['volume', 'open', 'high', 'low', 'close'], index=kiwoom.ohlcv['TR_code'])

    con = sqlite3.connect("C:/Users/82105/PycharmProjects/pythonProject/stock.db")
    df.to_sql('039490', con, if_exists='replace')
```

```python
class Kiwoom_Minute(QAxWidget):
    def __init__(self):
        super().__init__()
        self._create_kiwoom_instance()
        self._set_signal_slots()
        
    # KIWOOM OPEN API 실행
    def _create_kiwoom_instance(self):
        self.setControl("KHOPENAPI.KHOpenAPICtrl.1")

    # tr목록을 세팅
    def _set_signal_slots(self):
        self.OnEventConnect.connect(self._event_connect)
        self.OnReceiveTrData.connect(self._receive_tr_data)

    # 로그인이 되었는지 수행
    def comm_connect(self):
        self.dynamicCall("CommConnect()")
        self.login_event_loop = QEventLoop()
        self.login_event_loop.exec_()

    def _event_connect(self, err_code):
        if err_code == 0:
            print("connected")
        else:
            print("unconnected")

        self.login_event_loop.exit()
        
    # 종목코드 추출
    def get_code_list_by_market(self, market):
        code_list = self.dynamicCall("GetCodeListByMarket(QString)", market)
        code_list = code_list.split(';')
        return code_list[:-1]

    # 종목코드에 해당하는 주식회사를 한글로 입력
    def get_master_code_name(self, code):
        code_name = self.dynamicCall("GetMasterCodeName(QString)", code)
        return code_name

    def set_input_value(self, id, value):
        self.dynamicCall("SetInputValue(QString, QString)", id, value)

    # 데이터 추출
    def comm_rq_data(self, rqname, trcode, next, screen_no):
        self.dynamicCall("CommRqData(QString, QString, int, QString)", rqname, trcode, next, screen_no)
        self.tr_event_loop = QEventLoop()
        self.tr_event_loop.exec_()

    def _get_comm_data(self, trcode, record_name, index, item_name):
        ret = self.dynamicCall("GetCommData(QString, QString, int, QString)", trcode,
                               record_name, index, item_name)
        return ret.strip()

    def _get_repeat_cnt(self, trcode, rqname):
        ret = self.dynamicCall("GetRepeatCnt(QString, QString)", trcode, rqname)
        return ret

    def _receive_tr_data(self, screen_no, rqname, trcode, record_name, next, unused1, unused2, unused3, unused4):
        if next == '2':
            self.remained_data = True
        else:
            self.remained_data = False

        if rqname == "opt10080_req":
            self._opt10080(rqname, trcode)

        try:
            self.tr_event_loop.exit()
        except AttributeError:
            pass

    # 분봉 데이터들을 추출하여서 분석 및 저장
    def _opt10080(self, rqname, trcode):
        data_cnt = self._get_repeat_cnt(trcode, rqname)

        for i in range(data_cnt):
            TR_code = self._get_comm_data(trcode, rqname, i, "종목코드")
            volume = self._get_comm_data(trcode, rqname, i, "거래량")
            open = self._get_comm_data(trcode, rqname, i, "시가")
            high = self._get_comm_data(trcode, rqname, i, "고가")
            low = self._get_comm_data(trcode, rqname, i, "저가")
            close = self._get_comm_data(trcode, rqname, i, "현재가")
            print(TR_code, volume + "원" + open + "원" + high + "원" + low + "원" + close + "원")

            self.ohlcv['TR_code'].append(TR_code)
            self.ohlcv['volume'].append(int(volume))
            self.ohlcv['open'].append(int(open))
            self.ohlcv['high'].append(int(high))
            self.ohlcv['low'].append(int(low))
            self.ohlcv['close'].append(int(close))


if __name__ == "__main__":
    app = QApplication(sys.argv)
    kiwoom = Kiwoom_Minute()
    kiwoom.comm_connect()
    kiwoom.ohlcv = {'TR_code': [], 'volume': [], 'open': [], 'high': [], 'low': [], 'close': []}

    # opt10081 TR 요청
    kiwoom.set_input_value("종목코드", "039490")
    kiwoom.set_input_value("틱범위", "15")
    kiwoom.set_input_value("수정주가구분", 1)
    kiwoom.comm_rq_data("opt10080_req", "opt10080", 0, "0101")

    while kiwoom.remained_data == True:
        time.sleep(TR_REQ_TIME_INTERVAL)
        kiwoom.set_input_value("종목코드", "039490")
        kiwoom.set_input_value("틱범위", "15")
        kiwoom.set_input_value("수정주가구분", 1)
        kiwoom.comm_rq_data("opt10080_req", "opt10080", 2, "0101")
        
    # pandas를 이용하여서 DB Database SQLITE에다가 저장
    df = pd.DataFrame(kiwoom.ohlcv, columns=['volume', 'open', 'high', 'low', 'close'], index=kiwoom.ohlcv['TR_code'])

    con = sqlite3.connect("C:/Users/82105/PycharmProjects/pythonProject/stock.minute.db")
    df.to_sql('039490', con, if_exists='replace')
```
